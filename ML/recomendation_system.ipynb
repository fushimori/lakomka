```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Electronic_sales_Sep2023-Sep2024.csv")
df
```

```python
df = df[df['Order Status'] == 'Completed']
df
```

```python
df.isnull().sum()
```

```python
user_item_interactions = df.groupby(['Customer ID', 'SKU'])['Rating'].mean().reset_index()
user_item_interactions
```

```python
user_ids = user_item_interactions['Customer ID'].unique()
product_ids = user_item_interactions['SKU'].unique()

user_mapping = {user_id: idx for idx, user_id in enumerate(user_ids)}
product_mapping = {product_id: idx for idx, product_id in enumerate(product_ids)}

len(product_mapping)
```

```python
from scipy.sparse import csr_matrix

user_item_matrix = csr_matrix(
    (
        user_item_interactions['Rating'],  # Рейтинги как веса
        (
            user_item_interactions['Customer ID'].map(user_mapping),
            user_item_interactions['SKU'].map(product_mapping)
        )
    ),
    shape=(len(user_ids), len(product_ids))
)

print(user_item_matrix)
```

```python
from implicit.als import AlternatingLeastSquares

# Настройка модели
model = AlternatingLeastSquares(
    factors=50,  # Размерность скрытого пространства
    regularization=0.1,  # Регуляризация
    iterations=20  # Количество итераций
)

# Обучение модели
model.fit(user_item_matrix)
```

```python
def recommend_products(user_id, user_mapping, reverse_product_mapping, model, user_item_matrix, top_n=5):
    if user_id not in user_mapping:
        raise ValueError(f"Пользователь с ID {user_id} отсутствует в данных.")
    
    user_idx = user_mapping[user_id]
    
    if user_item_matrix[user_idx].nnz == 0:
        raise ValueError(f"Для пользователя с ID {user_id} нет взаимодействий.")
    
    # Получение рекомендаций
    recommendations = model.recommend(
        user_idx,
        user_item_matrix[user_idx],
        N=top_n,
        filter_already_liked_items=False
    )
    
    
    product_list = list(product_mapping.keys())

# Теперь сопоставим индексы из рекомендаций с реальными товарными идентификаторами (SKU)
    recommended_product_ids = [product_list[idx] for idx in recommendations[0]]
    return recommended_product_ids

```

```python
reverse_product_mapping = {v: k for k, v in product_mapping.items()}
user_id = 1002

if user_id not in user_mapping:
    raise ValueError(f"Пользователь с ID {user_id} отсутствует в маппинге.")
    
user_idx = user_mapping[user_id]

if user_item_matrix[user_idx].nnz == 0:  # Проверяем количество ненулевых элементов
    raise ValueError(f"Для пользователя с ID {user_id} нет взаимодействий.")


print(f"Размерность матрицы взаимодействий: {user_item_matrix.shape}")

recommended_products = recommend_products(
    user_id, 
    user_mapping, 
    reverse_product_mapping, 
    model, 
    user_item_matrix,
    top_n=10
)

print(f"Рекомендованные товары для пользователя {user_id}: {recommended_products}")
```

```python
def ndcg_at_k(y_true, y_pred, k):
    # Переводим y_true в массив релевантности

    k = min(len(y_true), k)
    y_true = [1 if item in y_pred[:k] else 0 for item in y_true]
    y_pred = y_pred[:k]
    
    dcg = np.sum((2 ** np.array(y_true) - 1) / np.log2(np.arange(2, len(y_pred) + 2)))
    
    # Для IDCG используем идеальное ранжирование
    ideal_relevance = sorted(y_true, reverse=True)
    idcg = np.sum((2 ** np.array(ideal_relevance) - 1) / np.log2(np.arange(2, len(ideal_relevance) + 2)))
    
    return dcg / idcg if idcg > 0 else 0
```

```python
def hit_rate(true_items, recommendations):
    """
    Вычисление метрики Hit Rate.
    
    :param true_items: Список релевантных товаров.
    :param recommendations: Список рекомендаций или одиночное значение.
    :return: 1, если хотя бы один релевантный элемент найден, иначе 0.
    """
    # Убедимся, что recommendations — это список
    if isinstance(recommendations, (int, np.integer)):
        recommendations = [recommendations]
    
    # Проверяем пересечение рекомендованных и релевантных элементов
    return int(any(item in true_items for item in recommendations))
```

```python
def mean_average_precision(y_true, y_pred, k):
    score = 0.0
    num_hits = 0.0
    for i, p in enumerate(y_pred[:k]):
        if p in y_true and p not in y_pred[:i]:
            num_hits += 1.0
            score += num_hits / (i + 1.0)
    return score / min(len(y_true), k) if len(y_true) > 0 else 0
```

```python
def evaluate_model(model, user_item_matrix, user_mapping, reverse_product_mapping, top_n=5):
    ndcg_scores = []
    hit_rate_scores = []
    map_scores = []

    for user_id, user_idx in user_mapping.items():
        if user_item_matrix[user_idx].nnz == 0:
            continue  # Пропускаем, если у пользователя нет взаимодействий в тесте
        
        # Получаем реальные товары из теста
        true_items = user_item_matrix[user_idx].indices
        # print(true_items)
        # Генерируем рекомендации для конкретного пользователя
        recommendations = model.recommend(
            user_idx, 
            user_item_matrix,  # Используем обучающую матрицу
            N=top_n,
            filter_already_liked_items=False
        )
        
        # Преобразуем индексы рекомендаций в SKU
        product_list = list(product_mapping.keys())

    # Теперь сопоставим индексы из рекомендаций с реальными товарными идентификаторами (SKU)
        recommended_product_ids = [product_list[idx] for idx in recommendations[0]]

        # Рассчитываем метрики
        ndcg_scores.append(ndcg_at_k(true_items, recommendations[0], top_n))
        hit_rate_scores.append(hit_rate(true_items, recommendations[0]))
        map_scores.append(mean_average_precision(true_items, recommendations[0], top_n))

    # Средние значения метрик
    return {
        "NDCG": np.mean(ndcg_scores),
        "Hit Rate": np.mean(hit_rate_scores),
        "MAP": np.mean(map_scores)
    }
```

```python
reverse_product_mapping = {v: k for k, v in product_mapping.items()}


metrics = evaluate_model(model, user_item_matrix, user_mapping, reverse_product_mapping, top_n=5)
print("Результаты модели:")
print(f"NDCG: {metrics['NDCG']:.4f}")
print(f"Hit Rate: {metrics['Hit Rate']:.4f}")
print(f"MAP: {metrics['MAP']:.4f}")
```
